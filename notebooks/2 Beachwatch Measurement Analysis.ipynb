```python
import matplotlib.pyplot as plt 
import metapack as mp
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pkg = mp.open_package('http://library.metatab.org/ceden.waterboards.ca.gov-beachwatch-sandiego-2.zip')

pkg

df = pkg.resource('beachwatch-sd').read_csv(parse_dates=True)

# It looks like the prefix of the station code groups stations, maybe into watersheds. 
df['stationgroup'] = df.stationcode.str[:2]

# The results has a large range, so log transformation makes them easier to visualize.
df['log_result'] = df.result.apply(np.log10)
```

```python
dfa = df[['sampledate','stationcode','analyte', 'result','methodname','resultqualcode','qacode']].copy()
```

```python
dfa['result'] = dfa['result'].replace([np.inf, -np.inf], np.nan)
dfa['l_result'] = np.log(dfa.result, where=(dfa.result > 0) ).replace([np.inf, -np.inf], np.nan)
#dfa['l_result_z'] = (dfa['l_result']-dfa['l_result'].mean())/dfa['l_result'].std()
```

```python
df[['sampledate','stationcode','analyte', 'result','methodname']].set_index(['sampledate','stationcode','analyte','methodname'])
```
